```python
#if,elif & else statement

x = int(input('Enter int values X = '))
y = int(input('Enter int values Y = '))
if x>y:
    print("{} is greater than {}".format(x,y))
elif x==y:
    print("{} is equal to {}".format(x,y))
else:          
    print("{} is leser than {}".format(x,y))
```
